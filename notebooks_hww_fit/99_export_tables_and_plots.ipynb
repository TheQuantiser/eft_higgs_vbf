```python
from pathlib import Path
import json, sys, os, math
import numpy as np
import pandas as pd

def find_repo_root(start=None, marker='smeft_contents.txt', max_up=8):
    p = Path(start or Path.cwd()).resolve()
    for _ in range(max_up+1):
        if (p / marker).exists():
            return p
        if p.parent == p:
            break
        p = p.parent
    raise FileNotFoundError(f'Could not find repo root containing {marker} from {Path.cwd()}')

REPO = find_repo_root()
NOTEBOOK_DIR = REPO / 'notebooks_hww_fit'
SMEFT_LISTING = REPO / 'smeft_contents.txt'
LOCAL_DICT = NOTEBOOK_DIR / 'dictionary.json'
print('repo:', REPO)
import matplotlib.pyplot as plt
OUT=REPO/'notebooks_hww_fit'/'outputs'
OUT.mkdir(exist_ok=True, parents=True)
print('output dir:',OUT)
```

```python
def export_df(df, name):
    csv=OUT/f'{name}.csv'
    pkl=OUT/f'{name}.pkl'
    df.to_csv(csv,index=False)
    df.to_pickle(pkl)
    print('wrote',csv,'and',pkl)

def export_json(obj,name):
    p=OUT/f'{name}.json'
    p.write_text(json.dumps(obj,indent=2,default=str))
    print('wrote',p)
```

```python
def save_current_fig(name):
    p=OUT/f'{name}.png'
    plt.savefig(p,dpi=150,bbox_inches='tight')
    print('wrote',p)

print('ready')
```
