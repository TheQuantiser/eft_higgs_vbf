# 99 Export all fit outputs + style-consistent plots

## Notebook purpose
Centralize export of numerical and visual outputs for reproducibility and downstream note preparation.

## Export conventions
- CSV/PKL for scan tables.
- JSON for metadata/summary objects.
- PNG for publication-style figures.

## Reproducibility tip
Always export both raw scan grids and summarized intervals so results can be revalidated without rerunning expensive scans.

```python
from pathlib import Path
import json, matplotlib.pyplot as plt, mplhep as hep
hep.style.use('CMS')
OUT=Path('outputs'); OUT.mkdir(exist_ok=True,parents=True)
```

```python
def export_df(df,name):
    df.to_csv(OUT/f'{name}.csv',index=False)
    df.to_pickle(OUT/f'{name}.pkl')

def export_json(obj,name):
    (OUT/f'{name}.json').write_text(json.dumps(obj,indent=2,default=str))

def save_fig(name):
    plt.savefig(OUT/f'{name}.png',dpi=150,bbox_inches='tight')
```
