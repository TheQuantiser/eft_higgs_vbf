```python
def values_1d_plot(coffea_name, operator_list):
    fig, ax = plt.subplots(figsize=(4 + len(operator_list), 5))

    for idx, op in enumerate(operator_list):
        wc_space_dict = sf.stxs_fit(coffea_name, [op])
        best_point, allowed_points, allowed_regions = sf.one_sigma_1d(wc_space_dict)
        twosig_best_point, twosig_allowed_points, twosig_allowed_regions = sf.two_sigma_1d(wc_space_dict)

        best_val = float(np.array(best_point))
        ax.plot(idx, best_val, 'o', color='tab:blue')

        #print(op, best_point)

        for i, two_sig_region in enumerate(twosig_allowed_regions):
            r_low = float(np.array(two_sig_region[0]))
            r_high = float(np.array(two_sig_region[-1]))
            ax.plot([idx, idx], [r_low, r_high], ':', color='tab:orange', linewidth=2,
                   label='2 sigma' if idx == 0 and i == 0 else None)
        
        for i, region in enumerate(allowed_regions):
            r_low = float(np.array(region[0]))
            r_high = float(np.array(region[-1]))
            ax.plot([idx, idx], [r_low, r_high], '-', color='tab:blue', linewidth=2,
                   label='1 sigma' if idx == 0 and i == 0 else None)
            #print(r_low, r_high)

    ax.axhline(0, color='tab:grey', linestyle=':', linewidth=0.8)
    ax.set_xticks(range(len(operator_list)))
    ax.set_xticklabels(operator_list)
    ax.set_ylabel("Wilson Coefficient value")
    ax.set_title("1D chi-squared fits, 1 & 2 sigma")
    plt.ylim(-10,10)
    #plt.yscale('symlog')
    plt.legend()
    plt.tight_layout()
    plt.show()

values_1d_plot("VBF_SMEFTsim_topU3l_NP1", 
         ['cHbox', 'cHDD', 'cHj1', 'cHj3', 'cHu', 'cHd', 'cHW', 'cHB', 'cHWB', 'cHWtil', 'cHBtil', 'cHWBtil'])
```

```python
def chi2_1d_plot(coffea_name, operator_list):
    plt.figure()

    for op in operator_list:
        wc_space_dict = sf.stxs_fit(coffea_name, [op])

        x = np.array(sorted(wc_space_dict.keys()), dtype=float)
        # for p in x:
        #     print(p, wc_space_dict[p]['total'][1])
        chi2 = np.array([wc_space_dict[p]['total'][1] for p in x], dtype=float)

        plt.plot(x, chi2, label=op)
    plt.ylim(0, 10)
    plt.ylabel(r'$\chi^2$')
    plt.xlabel("Wilson coefficient value")
    plt.xlim(-10,10)
    plt.title(rf"1D $\chi^2$ scan for {coffea_name}")
    plt.legend()
    plt.show()

chi2_1d_plot("VBF_SMEFTsim_topU3l_NP1", 
         ['cHW', 'cHWtil'])
```

```python
def chi2_2D_heatmap(coffea_name, operator_list):
    wc_space_dict = sf.stxs_fit(coffea_name, operator_list)

    c1_vals = sorted(set(p[0] for p in wc_space_dict))
    c2_vals = sorted(set(p[1] for p in wc_space_dict))

    X, Y = np.meshgrid(c1_vals, c2_vals)

    Z = np.zeros((len(c2_vals), len(c1_vals)))

    for i, c2 in enumerate(c2_vals):
        for j, c1 in enumerate(c1_vals):
            Z[i, j] = wc_space_dict[(c1, c2)]['total'][1]
    Z_plot = Z - np.min(Z)
    
    fig, ax = plt.subplots(figsize=(8, 6))

    cmap = plt.cm.viridis.copy()
    cmap.set_over('white')
    mesh = ax.pcolormesh(X, Y, Z_plot, cmap=cmap, vmin = 0, vmax = 10)
    cbar = plt.colorbar(mesh, ax=ax, label=r'$∆\chi^2$')
    cbar.set_ticks([0, 2.30, 5.99, 10])

    #1 sig and 2 sig contours
    one_sig_contour = ax.contour(X, Y, Z_plot, levels=[2.30], colors='red')
    ax.clabel(one_sig_contour, inline=False, fontsize=10)
    two_sig_contour = ax.contour(X, Y, Z_plot, levels=[5.99], colors='black', linestyles = 'dotted')
    ax.clabel(two_sig_contour, inline=False, fontsize=10)

    legend_elements = [
        Line2D([0], [0], color='red', lw=2, label=r'$1\sigma\ (\Delta\chi^2=2.30)$'),
        Line2D([0], [0], color='black', lw=2, linestyle='dotted', label=r'$2\sigma\ (\Delta\chi^2=5.99)$'),
        Line2D([0], [0], marker='o', color='red', linestyle='None', label='SM'),
        Line2D([0], [0], marker='x', color='red', linestyle='None', label='Best-fit'),
    ]
    
    ax.legend(handles=legend_elements)

    #SM point and best-fit point
    ax.scatter(0, 0, color='red', marker='o', s=10, label='SM')
    best_point, allowed_points, allowed_regions = sf.one_sigma_1d(wc_space_dict)
    xbest,ybest = best_point
    ax.scatter(xbest, ybest, color='red', marker='x', s=100, label='Best-fit')

    ax.set_xlim(-7, 7)
    ax.set_ylim(-7,7)
    ax.set_xlabel(operator_list[0])
    ax.set_ylabel(operator_list[1])
    ax.set_title(r'$∆\chi^2$ Likelihood scan')
    #ax.legend()

    plt.show()
    
chi2_2D_heatmap("VBF_SMEFTsim_topU3l_NP1", ['cHW', 'cHWtil'])
```

```python
def predicted_stxs_plot(coffea_name, operator):
    single_operator_list = [operator]
    wc_space_dict = sf.stxs_fit(coffea_name, single_operator_list)
    best_point, allowed_points, allowed_regions = sf.one_sigma_1d(wc_space_dict)
    
    best_point = None
    best_chisq = float('inf')
    
    for wcpoint, values in wc_space_dict.items():
        chisq = values['total'][1] 
        if chisq < best_chisq:
            best_chisq = chisq
            best_point = wcpoint

    bin1_xsec = wc_space_dict[best_point]['bin1'][0]
    bin2_xsec = wc_space_dict[best_point]['bin2'][0]
    total_chi = wc_space_dict[best_point]['total'][1]

    obs = np.array([240.0, 120.0])
    obs_err_up = np.array([200, 68])
    obs_err_down = np.array([190, 61])
    
    edges = np.array([1,2,3])
    centers = (edges[:-1] + edges[1:]) / 2
    
    fig, ax = plt.subplots(figsize=(5,4))
    
    hep.histplot(obs, bins=edges, histtype="barstep", linewidth=0, ax=ax, color='tab:green')
    hep.histplot([bin1_xsec, bin2_xsec], bins=edges, histtype="barstep", ax=ax, color = 'tab:blue',
                label=fr"SMEFTsim VBF, {operator} = {best_point:.3g}")
    
    ax.set_xticks([1.5, 2.5])
    ax.set_xticklabels(["bin1", "bin2"])
    ax.set_ylabel(r"$\sigma$")
    ax.set_yscale('log')
    
    ax.errorbar(centers,obs,yerr=[obs_err_up, obs_err_down],
        fmt="o",capsize=2,elinewidth=1,color = 'tab:orange',
        label="CMS data")
    
    # ax.errorbar(centers,VBF,yerr=[VBF_err_up, VBF_err_down],
    #     fmt="none",capsize=2,elinewidth=1,color = 'tab:blue',
    #     label=f"VBF_SMEFTsim_topU3l_NP1 ({bin1_xsec:.3g}:{bin2_xsec:.3g})")
    
    ax.legend(frameon=False)
    plt.title(fr'1D Best fit for {operator}')
    plt.tight_layout()
    plt.show()

predicted_stxs_plot('VBF_SMEFTsim_topU3l_NP1', 'cHWtil')
```

```python
def values_1d_plot(root_name, coffea_name, operator_list):

    results = sf.operator_sigma_1d(root_name, coffea_name, operator_list)
    
    ops = list(results.keys())
    best = np.array([float(results[op][0]) for op in ops])
    low = np.array([float(results[op][1]) for op in ops])
    high = np.array([float(results[op][2]) for op in ops])

    err_low = best - low
    err_high = high - best

    op_axis = np.arange(len(ops))

    plt.figure()
    plt.errorbar(op_axis, best, yerr=[err_low, err_high], fmt='o')

    plt.xticks(op_axis, ops)
    plt.ylabel("Wilson Coefficient value")
    plt.title("1D Best fits, 1 sigma errors")

    plt.axhline(0, linestyle='--')
    plt.ylim(-10, 10)
    plt.show()

values_1d_plot(
    "/uscms/home/azhou/nobackup/smeft/cmseft/generation/CMSSW_13_2_9/src/start0_040226.root", 
         "start0", 
         ['cHW', 'cHWtil'])
```
